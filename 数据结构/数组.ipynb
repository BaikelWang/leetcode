# 数组

## 排序
1. 冒泡排序

```python
class Bubble_sort:
    def sort(self, arr):
        n = len(arr)
        for i in range(n):
            for j in range(0, n-i-1):
                if arr[j] > arr[j+1]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
        return arr

array = [64, 34, 25, 12, 22, 11, 90]
bubble_sort = Bubble_sort()
sorted_array = bubble_sort.sort(array)
print("Sorted array is:", sorted_array)
```

2. 选择排序

```python
class Solution:
    def selectionSort(self, arr):
        n = len(arr)
        # 一共要n趟才能排好序
        for i in range(n-1):
            min_idx = i
            for j in range(i+1,n):
                if arr[j] < arr[min_idx]:
                    min_idx = j # 找到最小元素的索引
            # 交换找到的最小元素和当前位置元素
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
        return arr
        
sol = Solution()
array = [64, 25, 12, 22, 11]
sorted_array = sol.selectionSort(array)
print("Sorted array is:", sorted_array)
```
