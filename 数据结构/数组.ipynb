## 排序
1. 冒泡排序

```python
class Bubble_sort:
    def sort(self, arr):
        n = len(arr)
        for i in range(n):
            for j in range(0, n-i-1):
                if arr[j] > arr[j+1]:
                    arr[j], arr[j+1] = arr[j+1], arr[j]
        return arr

array = [64, 34, 25, 12, 22, 11, 90]
bubble_sort = Bubble_sort()
sorted_array = bubble_sort.sort(array)
print("Sorted array is:", sorted_array)
```

2. 选择排序

```python
class Solution:
    def selectionSort(self, arr):
        n = len(arr)
        # 一共要n趟才能排好序
        for i in range(n-1):
            min_idx = i
            for j in range(i+1,n):
                if arr[j] < arr[min_idx]:
                    min_idx = j # 找到最小元素的索引
            # 交换找到的最小元素和当前位置元素
            arr[i], arr[min_idx] = arr[min_idx], arr[i]
        return arr
        
sol = Solution()
array = [64, 25, 12, 22, 11]
sorted_array = sol.selectionSort(array)
print("Sorted array is:", sorted_array)
```

3. 插入排序

```python
class InsertSort:
    def sort(self, arr):
        n = len(arr)
        for i in range(1, n):
            # 取出当前元素
            temp = arr[i]
            # 在已排序的序列中从后向前扫描
            j = i
            # 找到插入位置
            while j > 0 and arr[j-1] > temp:
                # 元素后移
                arr[j] = arr[j-1]
                # 继续向前扫描
                j -= 1
            # 插入当前元素
            arr[j] = temp
        return arr
    
insert_sort = InsertSort()
array = [12, 11, 13, 5, 6]
sorted_array = insert_sort.sort(array)
print("Sorted array is:", sorted_array)
```

4. 希尔排序

```python
class ShellSort:
    def sort(self, arr):
        n = len(arr)
        # 初始增量设为数组长度的一半
        gap = n // 2
        while gap > 0:
            for i in range(gap, n):
                # 对每个增量分组进行插入排序
                temp = arr[i]
                j = i
                while j >= gap and arr[j - gap] > temp:
                    arr[j] = arr[j - gap]
                    j -= gap
                arr[j] = temp
            gap //= 2
        return arr
    
shell_sort = ShellSort()
array = [12, 34, 54, 2, 3]
sorted_array = shell_sort.sort(array)
print("Sorted array is:", sorted_array)
```

5. 归并排序

```python
class MergeSort:
    # 合并两个有序数组
    def merge(self, left, right):
        merged = []
        i = j = 0
        # 逐个比较两个数组的元素，按顺序放入合并后的数组
        while i < len(left) and j < len(right):
            if left[i] <= right[j]:
                # 左边数组的元素较小，放入合并数组
                merged.append(left[i])
                i += 1
            else:
                # 右边数组的元素较小，放入合并数组
                merged.append(right[j])
                j += 1
        merged.extend(left[i:])
        merged.extend(right[j:])
        return merged
    # 递归分割数组并排序
    def sort(self, arr):
        if len(arr) <= 1:
            return arr
        # 分割数组
        mid = len(arr) // 2 # 将数组从中间位置分为左右两个数组
        left = self.sort(arr[:mid]) # 递归将左子数组进行分解和排序
        right = self.sort(arr[mid:]) # 递归将右子数组进行分解和排序
        return self.merge(left, right) # 把当前数组组中有序子数组逐层向上，进行两两合并
    
    def sort_inplace(self, arr):
        return self.sort(arr)
    
merge_sort = MergeSort()
array = [38, 27, 43, 3, 9, 82, 10]
sorted_array = merge_sort.sort_inplace(array)
print("Sorted array is:", sorted_array)
    
```
